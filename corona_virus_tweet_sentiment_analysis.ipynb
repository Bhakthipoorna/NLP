```python
import re
import nltk
nltk.download('stopwords') # this download is not required in jupyter
from nltk.corpus import stopwords
nltk.download('wordnet') # this download is not required in jupyter
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
#from nltk.stem.porter import PorterStemmer
#ps=PorterStemmer()

corpus=[]
for i in range(len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['OriginalTweet'][i])  # removing all the special character except letters and with space
    review=review.lower() #lowering all the sentences
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]  # in the initial stage we used stemming but words are not meaningfull so, we are using lemmatising
    review= ' '.join(review) 
    corpus.append(review)
```
